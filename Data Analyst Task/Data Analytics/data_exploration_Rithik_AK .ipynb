```python
# Display unique values count for each column
unique_counts = {col: data[col].nunique() for col in data.columns}

for col, count in unique_counts.items():
    print(f"Count of unique values in column '{col}': {count}")
```

### Range of Dates 

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
min_date = data['Order Date'].min()
max_date = data['Order Date'].max()
print(f"The range of order dates is from {min_date.date()} to {max_date.date()}.")
```

### Year wise sales and profit

```python
#Display total sales, profit and the rate of sales and profit year-wise
data['Order Year'] = data['Order Date'].dt.year
yearly_summary = data.groupby('Order Year').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
yearly_summary['Sales Change (%)'] = yearly_summary['Sales'].pct_change() * 100
yearly_summary['Profit Change (%)'] = yearly_summary['Profit'].pct_change() * 100
yearly_summary['Sales'] = yearly_summary['Sales'].apply(lambda x: f"${x:,.2f}")
yearly_summary['Profit'] = yearly_summary['Profit'].apply(lambda x: f"${x:,.2f}")
yearly_summary['Sales Change (%)'] = yearly_summary['Sales Change (%)'].apply(lambda x: f"{x:+.2f}%")
yearly_summary['Profit Change (%)'] = yearly_summary['Profit Change (%)'].apply(lambda x: f"{x:+.2f}%")

table = yearly_summary.values.tolist()
headers = ['Order Year', 'Total Sales', 'Sales Change (%)', 'Total Profit', 'Profit Change (%)']
print(tabulate(table, headers=headers, tablefmt='pretty'))
```

### Values of some categorical columns

```python
columns = ['Ship Mode', 'Segment', 'Category', 'Order Priority']
#unique values in these categorical columns
for col in columns:
    values = data[col].value_counts()
    table = [[value, count] for value, count in values.items()]
    print(f"Unique values and counts for column '{col}':")
    print(tabulate(table, headers=[col, 'Count'], tablefmt='pretty'))
    print() 
```

### Year wise sales and profit for each segment

```python
segment_yearly_summary = data.groupby(['Order Year', 'Segment']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
segment_yearly_summary['Sales'] = segment_yearly_summary['Sales'].apply(lambda x: f"${x:,.2f}")
segment_yearly_summary['Profit'] = segment_yearly_summary['Profit'].apply(lambda x: f"${x:,.2f}")
years = segment_yearly_summary['Order Year'].unique()

for year in sorted(years):
    year_data = segment_yearly_summary[segment_yearly_summary['Order Year'] == year]
    table = year_data.values.tolist()
    headers = ['Segment', 'Total Sales', 'Total Profit']    
    print(f"Segment-wise Sales and Profit for Year {year}:")
    print(tabulate(table, headers=headers, tablefmt='grid'))
    print() 
```

### Year wise sales and profit for each category

```python
category_yearly_summary = data.groupby(['Order Year', 'Category']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
category_yearly_summary['Sales'] = category_yearly_summary['Sales'].apply(lambda x: f"${x:,.2f}")
category_yearly_summary['Profit'] = category_yearly_summary['Profit'].apply(lambda x: f"${x:,.2f}")
years = category_yearly_summary['Order Year'].unique()

for year in sorted(years):
    year_data = category_yearly_summary[category_yearly_summary['Order Year'] == year]
    table = year_data.values.tolist()
    headers = ['Category', 'Total Sales', 'Total Profit']    
    print(f"Category-wise Sales and Profit for Year {year}:")
    print(tabulate(table, headers=headers, tablefmt='grid'))
    print() 
```

### Year wise sales and profit for each order priority

```python
priority_yearly_summary = data.groupby(['Order Year', 'Order Priority']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
priority_yearly_summary['Sales'] = priority_yearly_summary['Sales'].apply(lambda x: f"${x:,.2f}")
priority_yearly_summary['Profit'] = priority_yearly_summary['Profit'].apply(lambda x: f"${x:,.2f}")
years = priority_yearly_summary['Order Year'].unique()

for year in sorted(years):
    year_data = priority_yearly_summary[priority_yearly_summary['Order Year'] == year]
    table = year_data.values.tolist()
    headers = ['Order Priority', 'Total Sales', 'Total Profit']    
    print(f"Order Priority-wise Sales and Profit for Year {year}:")
    print(tabulate(table, headers=headers, tablefmt='grid'))
    print()
```

# Data Visualisation

```python
#importing packages
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.subplots as sp
import plotly.express as px
import plotly.graph_objects as go
import math
```

```python
#Line chart for overall sales over all years
data['Order Month'] = data['Order Date'].dt.month
monthly_sales = data.groupby(['Order Year', 'Order Month']).agg({'Sales': 'sum'}).reset_index()

fig = go.Figure()
years = sorted(monthly_sales['Order Year'].unique())

for year in years:
    year_data = monthly_sales[monthly_sales['Order Year'] == year]
    fig.add_trace(go.Scatter(
        x=year_data['Order Month'],
        y=year_data['Sales'],
        mode='lines+markers',
        name=f'Sales - {year}',
        line=dict(width=2),
        marker=dict(size=8)
    ))

fig.update_layout(
    title_text='<b>Monthly Sales Over All Years</b>',
    title_font=dict(size=25, family="Liberation Serif", color="#5b5b5b"),
    xaxis_title='Month',
    yaxis_title='Sales',
    plot_bgcolor='white',
    xaxis=dict(tickvals=list(range(1, 13)), ticktext=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']),
    yaxis=dict(tickfont=dict(color='gray')),
    legend_title='Year',
    legend=dict(
        x=1, 
        y=1,
        traceorder='normal',
        orientation='v'
    ),
    width=600, 
    height=600, 
)

fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray', griddash='dot')
fig.update_xaxes(tickfont=dict(color='gray'))
fig.show()
```

```python
#Bar plot for total sales and profit for each category
category_sales = data.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'})

fig = make_subplots(rows=1, cols=2, subplot_titles=("Total Sales", "Total Profit"))
fig.add_trace(go.Bar(x=category_sales.index, y=category_sales['Sales'], name='Sales', marker_color='blue'), row=1, col=1)
fig.add_trace(go.Bar(x=category_sales.index, y=category_sales['Profit'], name='Profit', marker_color='green'), row=1, col=2)

fig.update_layout(
    title_text='<b>Total Sales and Profitability of Each Product Category</b>',
    title_font=dict(size=25, family="Liberation Serif", color="#5b5b5b"),
    plot_bgcolor='white',
    bargap=0.5,
    showlegend=False,
    width=700, 
    height=500,
)

fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray', griddash='dot', tickfont=dict(color='gray'))
fig.update_xaxes(tickfont=dict(color='gray'))
fig.show()
```

```python
#Pie chart for total sales for each category year-wise
yearly_category_sales = data.groupby(['Order Year', 'Category']).agg({'Sales': 'sum'}).reset_index()
years = sorted(yearly_category_sales['Order Year'].unique())

fig = sp.make_subplots(
    rows=1,
    cols=len(years),
    subplot_titles=[f'Year: {year}' for year in years],
    specs=[[{'type': 'pie'} for _ in years]],
    horizontal_spacing=0.05  
)
color_palette = px.colors.qualitative.Plotly

for i, year in enumerate(years):
    year_data = yearly_category_sales[yearly_category_sales['Order Year'] == year]    
    fig.add_trace(
        go.Pie(
            labels=year_data['Category'],
            values=year_data['Sales'],
            name=f'Sales Distribution - {year}',
            marker=dict(colors=color_palette[:len(year_data)]),
            textinfo='label+percent',
            hole=0.3,  
            showlegend=False 
        ),
        row=1, col=i + 1
    )

fig.update_layout(
    title_text='<b>Sales Distribution by Category for Each Year</b>',
    title_font=dict(size=25, family="Liberation Serif", color="#5b5b5b"),
    plot_bgcolor='white',
    height=600,  
    width=250 * len(years),  
    showlegend=True,  
    legend_title='Category',
    legend=dict(
        x=1,  
        y=1,
        traceorder='normal',
        orientation='v',
        title_font=dict(size=14, family="Liberation Serif"),
        font=dict(size=12, color='black')
    ),
    margin=dict(t=100, b=50, l=50, r=150)  
)

fig.update_traces(
    textfont=dict(size=14, color='black'),
    marker=dict(line=dict(color='white', width=1)) 
)
fig.show()
```

```python
#Bar plot for total sales for each sub-category for each category
category_subcategory_sales = data.groupby(['Category', 'Sub-Category']).agg({'Sales': 'sum'}).reset_index()
categories = sorted(category_subcategory_sales['Category'].unique())

n = len(categories)
cols = math.ceil(math.sqrt(n))
rows = math.ceil(n / cols)

fig = sp.make_subplots(
    rows=rows,
    cols=cols,
    subplot_titles=categories,
    specs=[[{'type': 'bar'} for _ in range(cols)] for _ in range(rows)],
    horizontal_spacing=0.1, 
    vertical_spacing=0.2    
)

for i, category in enumerate(categories):
    category_data = category_subcategory_sales[category_subcategory_sales['Category'] == category]
    
    fig.add_trace(
        go.Bar(
            x=category_data['Sub-Category'],
            y=category_data['Sales'],
            name=category,
            marker_color=color_palette[i % len(color_palette)],
            text=[f"${x:,.2f}" for x in category_data['Sales']],
            textposition='auto'
        ),
        row=(i // cols) + 1,
        col=(i % cols) + 1
    )

fig.update_layout(
    title_text='<b>Sales by Sub-Category for Each Category</b>',
    title_font=dict(size=25, family="Liberation Serif", color="#5b5b5b"),
    plot_bgcolor='white',
    height=900,  
    width=800,  
    showlegend=False, 
    margin=dict(t=100, b=50, l=50, r=50) 
)

fig.update_yaxes(
    showgrid=True,
    gridwidth=0.5,
    gridcolor='lightgray',
    griddash='dot',
    tickfont=dict(color='gray'),
    title='Sales'
)

fig.update_xaxes(
    tickfont=dict(color='gray'),
    tickangle=-45  
)

# Show the plot
fig.show()
```

```python
#Bar chart for total sales for each sub-category under categories year-wise
year_category_subcategory_sales = data.groupby(['Order Year', 'Category', 'Sub-Category']).agg({'Sales': 'sum'}).reset_index()

n = len(categories)
cols = math.ceil(math.sqrt(n))
rows = math.ceil(n / cols)

fig = sp.make_subplots(
    rows=rows,
    cols=cols,
    subplot_titles=[f'{category}' for category in categories],
    specs=[[{'type': 'bar'} for _ in range(cols)] for _ in range(rows)],
    horizontal_spacing=0.1,  
    vertical_spacing=0.15  
)

for i, category in enumerate(categories):
    category_data = year_category_subcategory_sales[year_category_subcategory_sales['Category'] == category]
    
    for j, year in enumerate(years):
        year_data = category_data[category_data['Order Year'] == year]
        fig.add_trace(
            go.Bar(
                x=year_data['Sub-Category'],
                y=year_data['Sales'],
                name=str(year),
                marker_color=color_palette[j % len(color_palette)],
                text=[f"${x:,.2f}" for x in year_data['Sales']],
                textposition='auto',
                showlegend=(i == 0)  
            ),
            row=(i // cols) + 1,
            col=(i % cols) + 1
        )

fig.update_layout(
    title_text='<b>Sales by Sub-Category for Each Category and Year</b>',
    title_font=dict(size=25, family="Liberation Serif", color="#5b5b5b"),
    plot_bgcolor='white',
    height=1000,
    width=1000, 
    showlegend=True,
    legend_title='Year',
    legend=dict(
        x=1.05,  
        y=0.5,
        traceorder='normal',
        orientation='v',
        title_font=dict(size=14, family="Liberation Serif"),
        font=dict(size=12, color='black'),
        bgcolor='rgba(255, 255, 255, 0.8)',  
        bordercolor='lightgray',
        borderwidth=1
    ),
    margin=dict(t=100, b=50, l=50, r=200)  
)

fig.update_yaxes(
    showgrid=True,
    gridwidth=0.5,
    gridcolor='lightgray',
    griddash='dot',
    tickfont=dict(color='gray'),
    title='Sales'
)

fig.update_xaxes(
    tickfont=dict(color='gray'),
    tickangle=-45  # Rotate x-axis labels for better readability
)
fig.show()
```
